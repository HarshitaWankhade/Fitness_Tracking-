```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data.csv")
```

```python
df.head(3)
```

```python
df=df.head(11)
```

```python
df.plot()
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
df.plot(kind="scatter", x="Duration", y="Calories")
plt.title("Scatter Plot")
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.show()
```

```python
df.plot(kind="hist")
plt.title("Histogram")
plt.show()
```

```python
df.plot(kind="area")
plt.title("Area Plot")
plt.show()
```

```python
df.plot(kind="bar")
plt.title("bar chart")
plt.show()
```

```python
df.plot(kind="bar",stacked=True)
plt.title("Stacked bar chart")
plt.show()
```

```python
df.plot(kind="barh")
plt.title("horizontal bar chart")
plt.show()
```

```python
df.plot(kind="box")
plt.title("box plot")
plt.show()
```

```python
df.plot(kind="hexbin",x="Maxpulse", y="Calories" ,gridsize=18)
plt.title("Hexagonal Binning Plot")
plt.show()
```

```python
df.plot(kind="kde")
plt.title("density estimate plot")
plt.show()
```

```python
df.plot(kind="pie",y="Duration",autopct='%1.1f%%')
plt.title("Pie Plot")
plt.show()
```

```python
df.plot()
plt.title("Line Plot")
plt.show()
plt.savefig("lineplot.png")
```
